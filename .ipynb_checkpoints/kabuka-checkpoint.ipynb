```python
days = 20
tickers = {
    'apple':'AAPL',
    'facebook': 'FB',
    'google': 'GOOGL',
    'microsoft':'MSFT',
    'netflix':'NFLX',
    'amazon':'AMZN'
}

def get_data(days, tickers):
    df = pd.DataFrame() #空のデータフレームを用意
    for company in tickers.keys():
#company = 'facebook'
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist]) #dfにhistをどんどん追加していく
    return df
```

```python
for company in tickers.keys():
    print(company)
```

```python
days = 20
tickers = {
    'apple':'AAPL',
    'facebook': 'FB',
    'google': 'GOOGL',
    'microsoft':'MSFT',
    'netflix':'NFLX',
    'amazon':'AMZN',
    'terumo':'TRUMY'
}

get_data(days, tickers)


#'stanley':'STAEF.PK'
#'JUKI':'JUKIY'
```

```python
dF = get_data(days, tickers)
dF
```

```python
companies = ['apple', 'facebook']
data = dF.loc[companies]
data
```

```python
data.sort_index()
```

```python
data = data.T.reset_index()
data.head()
```

```python
data = pd.melt(data,id_vars=['Date']).rename(
    columns = {'value': 'Stock Prices(USD)'}
)
data
```

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None),
        color='Name:N'
    )
)
```
